```python
#Import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
import requests
from config import api_key
from citipy import citipy
import random as random
```

```python
#Build url for api access

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
#Generate a list of random latitude and longitude values

random_lat = []
random_lng = []
for x in range(0,1700):
    
    random_lat.append(random.uniform(-90.00,91))
    random_lng.append(random.uniform(-180.00,181))
#print(random_lat)
#print(random_lng)
lat_lng_df = pd.DataFrame({"lat":random_lat,"lng":random_lng})
#lat_lng_df.head()
#lat_lng_df.info()
```

```python
#Use citipy to pull the name and country code of the city closest to lat and lon coordinates
#build lists with this information

city_code = []
city_name = []
country_code = []
for index, row in lat_lng_df.iterrows():
    city = citipy.nearest_city(row['lat'],row['lng'])
    city_code.append(city)
    city_name.append(city.city_name)
    country_code.append(city.country_code)
```

```python
#Create city dataframe with the information gathered about cities above
#Clean the data set by removing duplicates

city_df = lat_lng_df.copy()
city_df["city_name"] = city_name
city_df["country_code"] = country_code
print(len(city_df["city_name"]))
city_df_clean = city_df.drop_duplicates(subset="city_name")
print(len(city_df_clean["city_name"].unique()))
city_df_clean.head()
```

```python
#Pull current weather data for the cities in the list using the Open Weather API
#Store the information in lists

name_l = []
clouds_l = []
country_l = []
date_l = []
humidity_l = []
lat_l = []
lng_l = []
max_temp_l = []
wind_speed_l = []

for index, row in city_df_clean.iterrows():
    response = requests.get(f"{query_url}{row[2]},{row[3]}").json()
    try:
        print(f"Currently printing city id: {response['id']} city name: {response['name']}")
    except KeyError:
        try:
            print(f"Currently printing city id: none city name: {response['name']}")
        except KeyError:
            print(f"Currently printing city id: none city name: none")
    try:
        name_l.append(response["name"])
    except KeyError:
        name_l.append(np.nan)
    try:
        clouds_l.append(response["clouds"]["all"])
    except KeyError:
        clouds_l.append(np.nan)
    try:
        country_l.append(response["sys"]["country"])
    except KeyError:
        country_l.append(np.nan)
    try:
        date_l.append(response["dt"])
    except KeyError:
        date_l.append(np.nan)
    try:
        humidity_l.append(response["main"]["humidity"])
    except KeyError:
        humidity_l.append(np.nan)
    try:
        lat_l.append(response["coord"]["lat"])
    except KeyError:
        lat_l.append(np.nan)
    try:
        lng_l.append(response["coord"]["lon"])
    except KeyError:
        lng_l.append(np.nan)
    try:
        max_temp_l.append(response["main"]["temp_max"])
    except KeyError:
        max_temp_l.append(np.nan)
    try:
        wind_speed_l.append(response["wind"]["speed"])
    except KeyError:
        wind_speed_l.append(np.nan)
```

```python
#Create a summary data frame
#Save data frame to a csv file
#Drop NaN values

cities_summ_df = pd.DataFrame({"City":name_l,"Cloudiness":clouds_l,"Country":country_l,
                              "Date":date_l,"Humidity":humidity_l,"Lat":lat_l,
                              "Lng":lng_l,"Max Temp":max_temp_l,"Wind Speed":wind_speed_l})

cities_summ_df.to_csv("cities_summary_data.csv",index = False,header = True)


cities_summ_df_clean = cities_summ_df.copy()
cities_summ_df_clean = cities_summ_df_clean.dropna(how='any')
cities_summ_df_clean.head()
```

```python
#Plot Latitude vs. Max Temp
from datetime import datetime

date = datetime.fromtimestamp(cities_summ_df_clean['Date'][0])

plt.rc('axes',edgecolor='white')

fig1, ax1 = plt.subplots()

ax1.scatter(cities_summ_df_clean["Lat"],cities_summ_df_clean["Max Temp"],marker='o',edgecolors='black',s=30)
ax1.set_facecolor("xkcd:light gray")
ax1.grid(color='w')
ax1.set_axisbelow(True)
ax1.set_facecolor("xkcd:light gray")
ax1.set_xlim(-80,100)
ax1.set_ylim(-100,150)
ax1.set_xlabel("Latitude")
ax1.set_ylabel("Max Temperature (F)")
ax1.set_title(f"City Latitude vs. Max Temperature {date.strftime('(%m/%d/%y)')}")
plt.savefig("Images/max_temp.png")
plt.show()
```

```python
#Plot Latitude vs. Humidity

fig2, ax2 = plt.subplots()

ax2.scatter(cities_summ_df_clean["Lat"],cities_summ_df_clean["Humidity"],marker='o',edgecolors='black',s=30)
ax2.set_facecolor("xkcd:light gray")
ax2.grid(color='w')
ax2.set_axisbelow(True)
ax2.set_xlim(-80,100)
ax2.set_ylim(-100,120)
ax2.set_xlabel("Latitude")
ax2.set_ylabel("Humidity (%)")
ax2.set_title(f"City Latitude vs. Humidity {date.strftime('(%m/%d/%y)')}")
plt.savefig("Images/humidity.png")
plt.show()
```

```python
#Plot Latitude vs. Cloudiness

fig3, ax3 = plt.subplots()

ax3.scatter(cities_summ_df_clean["Lat"],cities_summ_df_clean["Cloudiness"],marker='o',edgecolors='black',s=30)
ax3.set_facecolor("xkcd:light gray")
ax3.grid(color='w')
ax3.set_axisbelow(True)
ax3.set_xlim(-80,100)
ax3.set_ylim(-20,120)
ax3.set_xlabel("Latitude")
ax3.set_ylabel("Cloudiness (%)")
ax3.set_title(f"City Latitude vs. Cloudiness {date.strftime('(%m/%d/%y)')}")
plt.savefig("Images/cloudiness.png")
plt.show()
```

```python
#Plot Latitude vs. Wind Speed

fig4, ax4 = plt.subplots()

ax4.scatter(cities_summ_df_clean["Lat"],cities_summ_df_clean["Wind Speed"],marker='o',edgecolors='black',s=30)
ax4.set_facecolor("xkcd:light gray")
ax4.grid(color='w')
ax4.set_axisbelow(True)
ax4.set_xlim(-80,100)
ax4.set_ylim(-5,60)
ax4.set_xlabel("Latitude")
ax4.set_ylabel("Wind Speed (mph)")
ax4.set_title(f"City Latitude vs. Wind Speed {date.strftime('(%m/%d/%y)')}")
plt.savefig("Images/wind_speed.png")
plt.show()
```
